```python
from collections import deque

class Process:
  def __init__(self, pid, burst_time):
    self.pid = pid
    self.btime = burst_time

def round_robin(processes, time_quantum):
    queue = deque(processes)
    time = 0

    while queue:
        process = queue.popleft()
        if process.btime > time_quantum:
          print(f" time  {time} : process {process.pid} : runs for {time_quantum}")
          time += time_quantum
          process.btime -= time_quantum
          queue.append(process)
        else:
          print(f" time  {time} : process {process.pid} : runs for {process.btime}")
          time += process.btime


processes = [Process(1, 10), Process(2, 4), Process(3, 6)]
time_quantum = 4
round_robin(processes,time_quantum)
```
